```python
# Dependencies and Setup
import pandas as pd

# Load files, read csv and store into Pandas DF
school_data = pd.read_csv("Resources/schools_complete.csv")
student_data = pd.read_csv("Resources/students_complete.csv")

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete

```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#DISTRICT SUMMARY TABLE
#Calculate the total number of schools
total_schools = len(school_data['school_name'].unique())
#Calculate the total number of students
total_num_of_stu = school_data['size'].sum()
#Calculate the total budget
total_budget = school_data['budget'].sum()
#Calculate the average math score
average_math_score = student_data['math_score'].mean()
#Calculate the average reading score
average_reading_score = student_data['reading_score'].mean()
#Calculate the percentage of students with a passing math score (70 or greater)
pass_math_df= school_data_complete.loc[school_data_complete['math_score']>=70, ['Student ID','math_score']]
tot_pass_math = pass_math_df.shape[0]
percentage_passing_math = tot_pass_math/total_num_of_stu
#Calculate the percentage of students with a passing reading score (70 or greater)
pass_reading_df = school_data_complete.loc[school_data_complete['reading_score']>=70, ['Student ID','reading_score']]
tot_pass_reading = pass_reading_df.shape[0]
percentage_passing_reading = tot_pass_reading/total_num_of_stu
#Calculate the percentage of students who passed math and reading (% Overall Passing)
merge_passing_df = pd.merge(pass_math_df,pass_reading_df, how="inner", on=["Student ID"])
tot_overall_pass = merge_passing_df.shape[0]
percentage_overall_pass = tot_overall_pass/total_num_of_stu
#Create a dataframe to hold the above results
district_summary = pd.DataFrame({"Total Schools": [total_schools], "Total Students": total_num_of_stu, 'Total Budget': total_budget, 'Average Math Score': average_math_score, 'Average Reading Score': average_reading_score, '% Passing Math': percentage_passing_math, '% Passing Reading': percentage_passing_reading, '% Overall Passing': percentage_overall_pass})
district_summary
#Optional: give the displayed data cleaner formatting
district_summary.style.format({"Total Students": "{:,}","Total Budget": "${:,.2f}", 
                       "Average Reading Score": "{:.0f}", 
                       "Average Math Score": "{:.0f}", 
                       "% Passing Math": "{:.2%}", 
                       "% Passing Reading": "{:.2%}", 
                       "% Overall Passing": "{:.2%}"})
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#SCHOOL SUMMARY TABLE
#School Name
school_gp = school_data_complete.groupby(['school_name'])
#School Type
sch_type = school_data.set_index(['school_name'])['type']
#Total Students
sch_students = school_data_complete['school_name'].value_counts()

#Total School Budget
tot_sch_budget = school_gp.mean()['budget']

#Per Student Budget
sch_stud_budget = (tot_sch_budget / sch_students)

#Average Math Score
sch_avg_math = school_gp.mean()['math_score']

#Average Reading Score
sch_avg_reading = school_gp.mean()['reading_score']

#Percentage Passing Math
math_pass_df = school_data_complete.loc[school_data_complete['math_score'] >= 70]
math_pass = math_pass_df['school_name'].value_counts()
math_pass_pct = math_pass / sch_students

#Percentage Passing Reading
read_pass_df = school_data_complete.loc[school_data_complete['reading_score'] >= 70]
read_pass = read_pass_df['school_name'].value_counts()
read_pass_pct = read_pass / sch_students

#Percentage Overall Passing (The percentage of students that passed math and reading
overall_pass_df= school_data_complete.loc[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)]
overall_pass = overall_pass_df['school_name'].value_counts()
overall_pass_pct = overall_pass / sch_students

#School Summary Dataframe 
school_summary = pd.DataFrame({'School Type': sch_type,
                                'No. of Students': sch_students,
                                'School Budget': tot_sch_budget,
                                'Budget per Student': sch_stud_budget,
                                'Average Math Score': sch_avg_math,
                                'Average Reading Score': sch_avg_reading,
                                '% Passing Math': math_pass_pct,
                                '% Passing Reading': read_pass_pct,
                                '% Passing Overall': overall_pass_pct
                              })
school_summary
```

## Top Performing Schools (By % Overall Passing)

```python
top_five = school_summary.sort_values(['% Passing Overall'], ascending = False)
top_five.head(5)
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#Math Scores by Grade
#Create a pandas series for each grade. Hint: use a conditional statement.
nine_scr = school_data_complete.loc[school_data_complete['grade'] == '9th',:]
ten_scr = school_data_complete.loc[school_data_complete['grade'] == '10th',:]
eleven_scr = school_data_complete.loc[school_data_complete['grade'] == '11th',:]
twelve_scr = school_data_complete.loc[school_data_complete['grade'] == '12th',:]

#Group each series by school
nine_gb = nine_scr.groupby('school_name')
ten_gb = ten_scr.groupby('school_name')
eleven_gb = eleven_scr.groupby('school_name')
twelve_gb = twelve_scr.groupby('school_name')

#Create series of students for Average Math Scores
nine_avg_math = nine_gb['math_score'].mean()
ten_avg_math = ten_gb['math_score'].mean()
eleven_avg_math= eleven_gb['math_score'].mean()
twelve_avg_math = twelve_gb['math_score'].mean()

#Combine the series into a dataframe
math_score_summary = pd.DataFrame({
                '9th Grade Math Score': nine_avg_math,
                '10th Grade Math Score': ten_avg_math,
                '11th Grade Math Score': eleven_avg_math,
                '12th Grade Math Score': twelve_avg_math
        })
math_score_summary
```
